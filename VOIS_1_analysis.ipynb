```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('ANALYSIS.xlsx')
```

```python
df
```

```python
df.columns
```

```python
# Task 1: Identify top-selling products
```

```python
top_selling_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
print("\nTop selling products:")
print(top_selling_products)
```

```python
# Task 2: Analyze profitability
```

```python
profitability = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)
print("\nProfitability by sub-category:")
print(profitability)
```

```python
# Task 3: Distribution of sales across different regions
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum()
print("\nDistribution of sales across different regions:")
print(sales_by_region)
```

```python
# Task 4: Calculate total sales and profit for each segment
```

```python
common_ship_modes = df['Ship Mode'].value_counts()
print("\nMost common ship modes:")
print(common_ship_modes)
```

```python
# Task 5: Analyze the relationship between quantity sold and discount
```

```python
quantity_discount_relationship = df.groupby("Quantity")["Discount"].mean()
print("\nAverage discount based on quantity sold:")
print(quantity_discount_relationship)
```

```python
# Task 6: Identify the correlation between sales and profit
```

```python
sales_profit_correlation = df["Sales"].corr(df["Profit"])
print("\nCorrelation between sales and profit:")
print(sales_profit_correlation)
```

```python
# Task 7: Identify the most profitable cities
```

```python
profitable_cities = df.groupby("City")["Profit"].sum().sort_values(ascending=False).head(10)
print("\nMost profitable cities:")
print(profitable_cities)
```

```python
# Task 8: Identify the products with the highest and lowest profit margins
```

```python
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
highest_profit_margin_product = df.loc[df['Profit Margin'].idxmax(), 'Sub-Category']
lowest_profit_margin_product = df.loc[df['Profit Margin'].idxmin(), 'Sub-Category']
print("\nProduct with the highest profit margin:", highest_profit_margin_product)
print("Product with the lowest profit margin:", lowest_profit_margin_product)
```

```python
# Task 9: Identify the most common combinations of categories and sub-categories in orders
```

```python
order_combinations = df.groupby(['Category', 'Sub-Category']).size().nlargest(5)
print("\nMost common combinations of categories and sub-categories in orders:")
print(order_combinations)
```

```python
# Task 10: Identify the top-selling categories in each region:
```

```python
top_selling_categories_per_region = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
top_selling_categories_per_region = top_selling_categories_per_region.sort_values(by=['Region', 'Sales'], ascending=[True, False]).groupby('Region').head(1)
print("\nTop selling categories in each region:")
print(top_selling_categories_per_region)
```

```python
#Task 11: Identify the top-selling categories in each region:
```

```python
top_selling_categories_per_region = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
top_selling_categories_per_region = top_selling_categories_per_region.sort_values(by=['Region', 'Sales'], ascending=[True, False]).groupby('Region').head(1)
print("\nTop selling categories in each region:")
print(top_selling_categories_per_region)
```

```python
# task 13:Analyze the distribution of sales and profit by customer segment:
```

```python
sales_profit_by_segment = df.groupby("Segment")[["Sales", "Profit"]].sum()
print("\nTotal sales and profit by segment:")
print(sales_profit_by_segment)
```

```python
# task 14 :Identify the correlation between sales and quantity sold:
```

```python
# task 15 : Identify the correlation between sales and quantity sold:
```

```python
#Task 16 : Visualize the distribution of sales across different regions using a pie chart:
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum()
plt.figure(figsize=(8, 8))
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Sales Across Different Regions")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Task 17: Visualize the top-selling products using a horizontal bar graph:
```

```python
top_selling_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_selling_products.plot(kind='barh', color='skyblue')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.title('Top Selling Products')
plt.show()
```

```python
#Task 18: Visualize the relationship between quantity sold and discount using a scatter plot:
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Quantity'], df['Discount'], alpha=0.5)
plt.title('Relationship between Quantity Sold and Discount')
plt.xlabel('Quantity Sold')
plt.ylabel('Discount')
plt.grid(True)
plt.show()
```

```python
# Task19: Visualize the distribution of profit margins using a histogram:
```

```python
# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Profit Margin'], bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribution of Profit Margins')
plt.xlabel('Profit Margin')
plt.ylabel('Frequency')
plt.show()
```

```python
#Task 21 :Visualize the distribution of discounts for different quantities sold using a box plot:
```
